```python
# 想在一个绘画框里同时展现多个图，这时就需要学会对绘画框进行划分
import matplotlib.pyplot as plt
```

```python
# 比如让折线图跟散点图在同一个绘画框展现

# 将 绘画框 进行 对象化
fig = plt.figure()

# 将p1定义为绘画框的子图，211表示将绘画框划分为2行1列，最后的1表示第一幅图
p1 = fig.add_subplot(211)
x=[1,2,3,4,5,6,7,8]
y=[2,1,3,5,2,6,12,7]
p1.plot(x,y)

#将p2定义为绘画框的子图，211表示将绘画框划分为2行1列，最后的2表示第二幅图
p2=fig.add_subplot(212)
a=[1,2]
b=[2,4]
p2.scatter(a,b)

plt.show()
```

```python
# 举一反三，画 2*2 的图
fig = plt.figure()

p1 = fig.add_subplot(221)
x=[1,2,3,4,5,6,7,8]
y=[2,1,3,5,2,6,12,7]
p1.plot(x,y)

p2=fig.add_subplot(222)
a=[1,2]
b=[2,4]
p2.scatter(a,b)

p3 = fig.add_subplot(223)
x=[1,2,3,8,5,6,7,8]
y=[2,1,3,6,2,6,12,7]
p3.plot(x,y)

p4=fig.add_subplot(224)
a=[1,3]
b=[3,4]
p4.scatter(a,b)

plt.show()
```
