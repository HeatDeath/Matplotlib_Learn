```python
# 如果我们想了解一些数据的趋势，将数据点跟点之间连成线是最直观的方法
import matplotlib.pyplot as plt
```

```python
# 定义所有点的 x 集合和 y 集合
x=[1,2,3,4,5,6,7,8]
y=[2,1,3,5,2,6,12,7]
```

```python
# 画一个折线图
plt.plot(x,y)
# 展示绘画框
plt.show()
```

```python
# 调整线的类型为 虚线，加上图 标签
plt.plot(x, y, linestyle='--', label='picture')
plt.show()
```
